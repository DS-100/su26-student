```python
# Method 1: Ignore rows where admitted is NaN
mask = df3["admitted"].notna()
rate_by_decile_1 = (
    df3[mask]
    .groupby("frl_percentile")[["admitted", "applied"]]
    .sum()
    .assign(admission_rate=lambda x: x["admitted"] / x["applied"])
)

sns.lineplot(rate_by_decile_1, x='frl_percentile', y='admission_rate').set(title='Ignore admitted=NaN')
```

```python
# Method 2: Impute NaN admitted with 0
df3_imputed = df3.assign(admitted_imp=df["admitted"].fillna(0))
rate_by_decile_2 = (
    df3_imputed.groupby("frl_percentile")[["admitted_imp", "applied"]]
    .sum()
    .assign(admission_rate=lambda x: x["admitted_imp"] / x["applied"])
)

sns.lineplot(rate_by_decile_2, x='frl_percentile', y='admission_rate').set(title='Replace admitted=NaN with 0')
```

```python
# Method 3: Impute NaN admitted with 2
df3_imputed = df3.assign(admitted_imp=df["admitted"].fillna(2))
rate_by_decile_3 = (
    df3_imputed.groupby("frl_percentile")[["admitted_imp", "applied"]]
    .sum()
    .assign(admission_rate=lambda x: x["admitted_imp"] / x["applied"])
)

sns.lineplot(rate_by_decile_3, x='frl_percentile', y='admission_rate').set(title='Replace admitted=NaN with 2')
```

```python
# Method 4: Impute NaN admitted with random int in [0, min(applied, 4)]
rng2 = np.random.default_rng(seed=7342)

def random_impute_decile(row):
    if pd.isna(row["admitted"]):
        return rng2.integers(0, 3)
    return row["admitted"]

df3_imputed = df3.assign(admitted_imp=df.apply(random_impute_decile, axis=1))
rate_by_decile_4 = (
    df3_imputed.groupby("frl_percentile")[["admitted_imp", "applied"]]
    .sum()
    .assign(admission_rate=lambda x: x["admitted_imp"] / x["applied"])
)

sns.lineplot(rate_by_decile_4, x='frl_percentile', y='admission_rate').set(title='Replace admitted=NaN with random 0 to 2')
```

```python
# Repeat of Method 4 for attendance rates
# But also ignore rows with admitted=NaN and attended=NaN

rng2 = np.random.default_rng(seed=7342)

def random_impute_decile(row):
    if pd.isna(row["attended"]):
        return rng2.integers(0, 3)
    return row["attended"]

mask = df3["admitted"].notna()
masked_df = df3[mask]
    
df3_imputed_filtered = masked_df.assign(attended_imp=masked_df.apply(random_impute_decile, axis=1))

rate_by_decile_4 = (
    df3_imputed_filtered.groupby("frl_percentile")[["attended_imp", "admitted"]]
    .sum()
    .assign(attendance_rate=lambda x: x["attended_imp"] / x["admitted"])
)

sns.lineplot(rate_by_decile_4, x='frl_percentile', y='attendance_rate').set(title='Replace attended=NaN with random 0 to 2')
```

```python
df3['dist_percentile_raw'] = df3['dist_to_ucb_miles'].rank(pct=True)

bins = [0, 0.25, 0.5, 0.75, 1]
labels = ['0-25', '25-50', '50-75', '75-100']

df3['dist_percentile'] = pd.cut(df3['dist_percentile_raw'], bins=bins, labels=labels, include_lowest=True)
df3
```

```python
df3_grouped = df3.groupby(['frl_percentile', 'dist_percentile'])[['applied', 'grade_12']].agg('sum')

df3_grouped['app_rate'] = df3_grouped['applied'] / df3_grouped['grade_12']

df3_grouped
```

```python
# Build filtered dataset
df3_scatter = df3.loc[
    (df3["tot_enrolled"] >= 100) & (df3["is_charter"] == "N")
].copy()

df3_scatter["dist_to_ucb_bucket"] = pd.qcut(
    df3_scatter["dist_to_ucb_miles"],
    q=5,
    labels=["1st (0–20%)", "2nd (20–40%)", "3rd (40–60%)", "4th (60–80%)", "5th (80–100%)"],
)
df3_scatter["app_rate"] = df3_scatter["applied"] / df3_scatter["grade_12"]

print(f"Rows after filtering: {len(df3_scatter)}")
df3_scatter["dist_to_ucb_bucket"].value_counts().sort_index()
```

```python
ordinal = ["1st", "2nd", "3rd", "4th", "5th"]
plot_df = df3_scatter.dropna(subset=["pct_free_reduced", "app_rate"])
plot_df["dist_to_ucb_bucket"] = plot_df["dist_to_ucb_bucket"].cat.rename_categories(
    {old: f"{o} dist_to_ucb quintile" for old, o in zip(plot_df["dist_to_ucb_bucket"].cat.categories, ordinal)}
)

g = sns.FacetGrid(plot_df, col="dist_to_ucb_bucket", col_wrap=3, sharex=False)
g.map_dataframe(sns.scatterplot, x="pct_free_reduced", y="app_rate", size="grade_12")
g.set_axis_labels("free_reduced_percentage", "app_rate")
g.set_titles(col_template="{col_name}")
g.add_legend(title="grade_12")
```
