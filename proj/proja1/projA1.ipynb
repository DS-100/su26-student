### 🔖 Sale Price
We begin by examining the distribution of our target variable `Sale Price`. We have provided the following helper method `plot_distribution` that you can use to visualize the distribution of the `Sale Price` using both the histogram and the box plot using a shared x-axis. Run the following 2 cells.

````python
def plot_distribution(data, label):
    fig, axs = plt.subplots(nrows=2)

    sns.distplot(
        data[label], 
        ax=axs[0]
    )
    sns.boxplot(
        x=data[label],
        width=0.3, 
        ax=axs[1],
        showfliers=False,
    )

    # Align axes
    spacer = np.max(data[label]) * 0.05
    xmin = np.min(data[label]) - spacer
    xmax = np.max(data[label]) + spacer
    axs[0].set_xlim((xmin, xmax))
    axs[1].set_xlim((xmin, xmax))

    # Remove some axis text
    axs[0].xaxis.set_visible(False)
    axs[0].yaxis.set_visible(False)
    axs[1].yaxis.set_visible(False)

    # Put the two plots together
    plt.subplots_adjust(hspace=0)
    fig.suptitle("Distribution of " + label)
````

````python
plot_distribution(initial_data, label='Sale Price')
````

We can see that the data is highly skewed to the right currently, probably because of some extremely expensive houses. To get a clearer view of the distribution, run the following cell to examine a more limited range of house prices from [\\$0, \\$1,000,000].

Continuing from the previous part, as you explore the dataset, you might still run into more outliers that prevent you from creating a clear visualization or capturing the trend of the majority of the houses. 

Below, we defined a function `remove_outliers` that removes outliers from the dataset based on a threshold value of a variable. For example, `remove_outliers(training_data, 'Building Square Feet', lower=500, upper=8000)` should return a copy of `data` with only observations that satisfy `Building Square Feet` less than 8000 (exclusive) and `Building Square Feet` greater than 500 (exclusive).

````python
def remove_outliers(data, variable, lower=-np.inf, upper=np.inf):
    """
    Input:
      data (DataFrame): the table to be filtered
      variable (string): the column with numerical outliers
      lower (numeric): observations with values lower than or equal to this will be removed
      upper (numeric): observations with values higher than or equal to this will be removed
    
    Output:
      a DataFrame with outliers removed
      
    Note: This function should not change or mutate the contents of data.
    """  

    return data[(data[variable] < upper) & (data[variable] > lower)]
````

One way to address these issues is to transform the data to a different scale, as we did with `Sale Price`. Another way is to remove outliers. 

Here, we define an outlier as any point that is 1.5 interquartile ranges above Q3 or below Q1 (exclusive of the boundary).

If you need a refresher on how the IQR works, you can read this [Wikipedia page on IQR](https://en.wikipedia.org/wiki/Interquartile_range).

````python
IQR = training_data['Estimate (Land)'].quantile(0.75) - training_data['Estimate (Land)'].quantile(0.25) 

upper_boundary = training_data['Estimate (Land)'].quantile(0.75) + (IQR * 1.5) 
lower_boundary = training_data['Estimate (Land)'].quantile(0.25) - (IQR * 1.5) 

filtered_training_data = remove_outliers(training_data, "Estimate (Land)", lower = lower_boundary, upper = upper_boundary)
````

#### An Important Note on One-Hot-Encoding 

Unfortunately, simply replacing the integers with the appropriate strings isn’t sufficient for using `Wall Material` in our model.  Since `Wall Material` is a categorical variable, we will have to one-hot-encode the data. [For more information on why we want to use one-hot-encoding, refer to this link](https://machinelearningmastery.com/why-one-hot-encode-data-in-machine-learning/).

We have provided a prebuilt ohe_wall_material function below. It takes a DataFrame containing the labeled `Wall Material` column and returns a new DataFrame with all original columns plus one binary indicator column for each wall material category. The new columns follow the naming format Wall `Material_MATERIAL`. Run the cell below to define and apply the function.

````python
from sklearn.preprocessing import OneHotEncoder

def ohe_wall_material(data):
    """
    One-hot-encodes wall material. New columns are of the form "Wall Material_MATERIAL".
    """
    
    oh_enc = OneHotEncoder()
    oh_enc.fit(data[['Wall Material']])
    oh_enc_columns = pd.DataFrame(oh_enc.transform(data[['Wall Material']]).toarray(), # .todense() also works
                           columns=oh_enc.get_feature_names_out(),
                           index = data.index)
    
    return data.merge(oh_enc_columns, left_index=True, right_index=True)
    

training_data_ohe = ohe_wall_material(training_data_mapped)
# This line of code will display only the one-hot-encoded columns in training_data_ohe that 
# have names that begin with “Wall Material_" 
training_data_ohe.filter(regex='^Wall Material_').head(10)
````
