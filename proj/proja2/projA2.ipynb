> If you are wondering why we are predicting **log(sale price)** instead of the sale price itself: In Project A1, we added the `Log Sale Price` column, which is the logarithm of the sale price column, to neutralize the impact of extreme right-skewed outliers on the fitted model. 
>
> However, even if your model fits well by accurately predicting the log sale price, that does **not** mean it will always perform well after exponentiating the prediction back to actual sale price. Here's a simple example to illustrate this:
>
> Let’s say the dataset of actual sale prices is { $e^0 = \$1$, $e^{10} \approx \$22,000$ }. In other words, we have one home worth $\$1$, and another home worth $\$22,000$.
>
> Suppose we train a constant model on log prices using MSE. The log prices are { $\log(e^0) = 0$, $\log(e^{10}) = 10$}.
>
> The model that would best predict the log prices {0, 10} will always output a log sale price of 5 (i.e., the average of 0 and 10). In other words, the model will always output a raw price of $e^5 \approx \$148$.
>
> Even though this model has minimized error on **log** prices, it has not minimized error on **actual** prices. If we predict $\$148$, we have a huge error on the home worth $\$22,000$!
>
> If we wanted to minimize error on actual prices, we should train the model on actual prices. If we were to train a constant model on actual prices, it would always predict $\frac{1+22000}{2} \approx 11000$.
>
> For the sake of accurately evaluating your model, we will be testing your predictions against the **log** of sale prices, rather than exponentiating them and comparing to the raw sale prices.
>
> That being said, we can still convert the predicted log sale price to the sale price to gain some intuition about the model and dataset, which we will explore in later questions.

Let's make one small tweak to our approach. Rather than generate the X variable for our model using two lines of code floating around on their own, let's generate the X variable using a function:

```
def model_one_pipeline(df):
    training_set_with_total_bedrooms = add_total_bedrooms(df)

    return X_m1
```

This will be useful for a couple of reasons:
1. It will give us a standard code pattern for creating models.
2. It will allow us to apply this function to the training (for fitting) and validation set (for evaluating).

For example, consider the code below which uses this function to fit a model then uses it again to compute the MSE on the validation set.

````python
def model_one_pipeline(df):
    # Create new DataFrame by applying add_total_bedrooms to df, our input DataFrame
    training_set_with_total_bedrooms = add_total_bedrooms(df)
    # Extract only feature data from our new DataFrame
    X_m1 = training_set_with_total_bedrooms[["Bedrooms"]]
    return X_m1

# Apply our newly created pipeline function to the train DataFrame to create our design matrix containing the “Bedroom” feature
X_train_m1 = model_one_pipeline(train)
# Create a LinearRegression object and assign it to model_m1
model_m1 = LinearRegression()
# Fit our new LinearRegression object to our design matrix
model_m1.fit(X_train_m1, train["Log Sale Price"])
# Apply our pipeline function to the original validation set in order to extract the “Bedroom” feature our model uses for prediction
X_validation_m1 = model_one_pipeline(validation)
# Use our transformed validation set to generate Log Sale Price predictions, and then find MSE for these predictions compared to actual Log Sale Price for the validation set
mean_squared_error(model_m1.predict(X_validation_m1), validation["Log Sale Price"])
````

````python
def rmse(predicted, actual):
    """
    Calculates RMSE from actual and predicted values.
    Input:
      predicted (1D array): Vector of predicted/fitted values
      actual (1D array): Vector of actual values
    Output:
      A float, the RMSE value.
    """
    return np.sqrt(np.mean((actual - predicted)**2))
````

````python
from sklearn.model_selection import KFold

def compute_CV_error(X_train, Y_train, folds=10):
    """
    Split the training data into `k` subsets.
    For each subset, 
        - Fit a model holding out that subset.
        - Compute the RMSE on that subset (the validation set).
    You should be fitting `k` models in total.
    Return a list of `k` RMSEs.

    Args:
        X_train (DataFrame): Training data.
        Y_train (Series): Label.
        folds (int): Number of cross-validation folds.
    
    Return:
         A list of `k` RMSEs.
    """
    model = lm.LinearRegression(fit_intercept=True) # Use model that Question 5 requires.
    kf = KFold(n_splits=folds)
    validation_errors = []
    
    for train_idx, valid_idx in kf.split(X_train):



        # 1. Write your explanation here.
        split_X_train, split_X_valid = X_train.iloc[train_idx], X_train.iloc[valid_idx]
        split_Y_train, split_Y_valid = Y_train.iloc[train_idx], Y_train.iloc[valid_idx]
      
      
        # 2. Write your explanation here.
        model.fit(split_X_train, split_Y_train)

        # 3. Write your explanation here.
        error = rmse(model.predict(split_X_valid), split_Y_valid)


        validation_errors.append(error)

    return validation_errors
    
````

````python
# Define any additional helper functions or variables you need here

def ohe(data, labels):
    """
    One-hot encode the specified columns.
    """
    assert isinstance(labels, list), "labels must be a list of columns"

    encoder = OneHotEncoder()
    encoder.fit(data[labels])

    dummies = pd.DataFrame(
        encoder.transform(data[labels]).todense(),
        columns=encoder.get_feature_names_out(),
        index=data.index
    )

    return data.join(dummies), encoder.get_feature_names_out()

def bin_col(data, colname, granularity):
    """
    Put a numerical column into categorical bins.
    """
    data = data.copy()

    min_value = data[colname].min()

    data[colname + " Bin"] = (
        (data[colname] - min_value) // granularity
    ).astype(str)

    return data
````

## Question 5d: Defining The Pipeline Function

Just as in the guided model from `q3b`, you'll be writing a function that generates features for your model. Your function should be called `final_model_pipeline` in the following cell. It should be just like `model_one_pipeline` and `model_two_pipeline` in that it takes as input a training or validation set, and outputs the appropriate columns to be provided to a model.

Note: The `final_model_pipeline` function starts with special code that ignores house prices above a fixed threshold. We've chosen a threshold of 1,000,000 dollars. Note that we already discarded all of the one dollar houses at the very beginning of the notebook, since this was just a placeholder value, so we don't need to remove these again. The reason we're setting this upper limit is that there are a small number of very expensive properties. Since the MSE is the squared difference between prediction and actual value, this means these properties will have a relatively large impact on the loss, and thus our model is going to try too hard to make these properties' predictions accurate. Since the model pays so much attention to only the most expensive houses, it will do worse on the much more common lower priced household.

You're welcome to experiment with a different threshold. 

To get a really rough sense of the quality of your final_model_pipeline, run the cell immediately below, which will print the RMSE on the training set. You should be able to get your RMSE on the training set below 0.65 on the Log Sale Price scale.

````python
from feature_func import remove_outliers

# Please include all of your feature engineering processes inside this function.
# Do not modify the parameters of this function.
def final_model_pipeline(df): 
    # make a copy to avoid modifying the original dataframe
    data = df.copy()

    # We can see the sale price so this must be a training set. Test sets will not have
    # the Sale Price column available. Let's avoid fitting house prices that are either
    # very low or very high. Do not remove this code, as keeping all the one dollar houses
    # and super expensive houses will affect your model negatively.
    if "Sale Price" in data.columns:
        data = remove_outliers(data, 
                               variable='Sale Price',
                               upper=1_000_000)

        # drop the sale price if present to avoid accidentally using these in the model
        data = data.drop(["Sale Price", "Log Sale Price"], axis='columns')
    
````

````python
# Run this cell before you get to next step.
def make_feature_engine_adapter(pipeline):
    """
    Adapt final_model_pipeline(df) -> X to the A2 interface:

        feature_engine_final(train_data) -> (X, y)
        feature_engine_final(test_data, is_test_set=True) -> X
    """
    training_columns = None

    def feature_engine_final(data, is_test_set=False):
        nonlocal training_columns

        working_data = data.copy()

        if is_test_set:

            pipeline_input = working_data.drop(
                columns=["Sale Price", "Log Sale Price"],
                errors="ignore"
            )

            X = pipeline(pipeline_input)


            if len(X) != len(working_data):
                raise ValueError(
                    "The pipeline removed rows from validation/test data."
                )

            if not X.index.equals(working_data.index):
                raise ValueError(
                    "The pipeline changed or reordered validation/test indices."
                )

            if training_columns is None:
                raise RuntimeError(
                    "Call feature_engine_final on training data before "
                    "processing validation/test data."
                )


            X = X.reindex(columns=training_columns, fill_value=0)

            return X


        if "Sale Price" not in working_data.columns:
            raise ValueError(
                "Training data must contain a Sale Price column."
            )


        working_data = working_data.loc[
            working_data["Sale Price"] > 499
        ].copy()


        working_data["Log Sale Price"] = np.log(
            working_data["Sale Price"]
        )


        y = working_data["Log Sale Price"].copy()
        X = pipeline(working_data)


        y = y.loc[X.index]

        if not X.index.equals(y.index):
            raise ValueError("X and y are not aligned.")

        training_columns = X.columns.copy()

        return X, y

    return feature_engine_final


feature_engine_final = make_feature_engine_adapter(
    final_model_pipeline
)
````

````python
def rmse(predicted, actual):
    """
    Calculates RMSE from actual and predicted values.
    Input:
      predicted (1D array): Vector of predicted/fitted values
      actual (1D array): Vector of actual values
    Output:
      A float, the RMSE value.
    """
    return np.sqrt(np.mean((actual - predicted)**2))
````

### Question 8a

As discussed in Project A1, RMSE—while a widely used and powerful error metric introduced in earlier lectures—is not always the most appropriate choice when evaluating the *fairness* of a property appraisal system. In Question 7, you already encountered some of RMSE's limitations, particularly its tendency to disproportionately emphasize errors in high-priced properties due to the squaring of residuals.

In this question, rather than relying on RMSE, we will train and evaluate our model using another custom fairness metric. Specifically, we will examine the **Mean Absolute Percentage Error (MAPE)**, which measures the average error as a percentage of the true value. This allows us to better assess whether the model is making relatively fair predictions across different segments of the housing market.

In the code cells that follow, we’ll explore how MAPE varies across price ranges—comparing the model’s relative performance on inexpensive versus expensive housing. This helps identify whether the model systematically favors or disadvantages certain groups of properties based on their price.

**Note:** You'll notice that we’re no longer using `lm.LinearRegression()`, but instead have started using `minimize`. This is because `scikit-learn` does not allow customization of the loss function in their standard linear regression model. The approach shown below provides an equivalent way to train a linear regression model using a custom error metric.

**Warning:** These cells take quite a long time to run. Please be patient, wait, and avoid restarting the kernel or rerunning these cells more than necessary.

````python
data = pd.read_csv("cook_county_train.csv", index_col='Unnamed: 0')
trainX, trainY = feature_engine_final(data)
# DO NOT MODIFY ANYTHING IN THIS CELL
np.random.seed(1337)
rows = np.random.choice(range(trainX.shape[0]), 3000, replace = False)
trainX = trainX.iloc[rows]
trainY = trainY.iloc[rows]
````

````python
# X is the design matrix (including bias column), y is the vector of true outputs, theta is the parameter vector
def mape(theta, X, y):
    y_pred = X @ theta  # compute predicted values using linear combination of features
    percentage_error = np.abs((y - y_pred) / y)  # calculate element-wise percentage errors
    return np.mean(percentage_error)  # return the MAPE
````

````python
from scipy.optimize import minimize

# Add bias (intercept) column to the design matrix
trainX_with_bias = np.column_stack([np.ones(trainX.shape[0]), trainX])

# Initialize parameter vector with zeros
theta_0 = np.zeros(trainX_with_bias.shape[1])

# Use scipy's minimize to find weights that minimize MAPE
res = minimize(mape, theta_0, args=(trainX_with_bias, trainY), method='Nelder-Mead')

# Optimal weights after training
theta_opt = res.x

theta_opt
````

````python
new_preds_df = pd.DataFrame({
    'True Log Sale Price'     : trainY,
    'Predicted Log Sale Price': trainX_with_bias @ theta_opt,
    'True Sale Price'         : np.e ** trainY,
    'Predicted Sale Price'    : np.e ** (trainX_with_bias @ theta_opt)
})

plt.figure(figsize=(8, 5))
plt.subplot(1, 2, 1)

mape_values = []
for i in np.arange(8, 14, 0.5):
    mape_values.append(mape_interval(new_preds_df, i, i + 0.5))

plt.bar(x=np.arange(8.25, 14.25, 0.5), height=mape_values, edgecolor='black', width=0.5)
plt.title('MAPE of Log Sale Price Across\nLog Sale Price Intervals', fontsize=10)
plt.xlabel('Log Sale Price')
plt.ylabel('Mean Absolute Percentage Error (MAPE)')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10);
````
